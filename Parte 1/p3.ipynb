```python
import pandas as pd
```

```python
business = pd.read_csv("business.csv", usecols=["city", "name" , "stars"])
business.dropna(inplace=True)
business.shape
```

```python
business.head(20)
```

vi que algunas ciudades estaban en minuscula, hago caso gral que este todo en minisculas

```python
business["city"] = business["city"].apply(lambda x: x.lower())
```

```python
business["city"]
```

pruebo sacando los pueblos con menos de 10 visitas

```python
grouped = business.groupby(["city"]).agg({"city": ['count']})
```

```python
lvl0 = grouped.columns.get_level_values(0)
lvl1 = grouped.columns.get_level_values(1)
grouped.columns = lvl0 + "_" + lvl1
grouped
```

```python
grouped.reset_index(inplace=True)
grouped
```

```python
ciudadesConMasDeDiez = grouped[grouped["city_count"] >= 10]
ciudadesConMasDeDiez
```

aplico el filtro del mean a las ciudades elegidas

```python
ciudadesConMasDeDiez = business[business["city"].isin(ciudadesConMasDeDiez["city"])]
ciudadesConMasDeDiez
```

Una vez filtrada las ciudades con menos de 10 reviews, ahora agrupo las ciudades y le agrego el mean de las stars

```python
grouped = ciudadesConMasDeDiez.groupby("city").agg({"stars":"mean"})
grouped
```

decidi redondear los valores a 2 cifras, considero un valor 3.615385 que sea igual a 3.615654

```python
grouped["stars"] = grouped["stars"].apply(lambda x : round(x,2))
```
