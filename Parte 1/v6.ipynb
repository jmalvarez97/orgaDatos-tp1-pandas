Una vez que tengo las 5 cadenas, concateno a todos en un solo df

```python
dfCompleto = pd.concat([mcDonalds_short, subway_short, tacoBell_short,  burgerKing_short, wendys_short])
```

```python
dfCompleto = dfCompleto.reset_index()[["name","business_id"]]
dfCompleto
```

Ahoro cargo las reviews, me quedo solo con el business_id y sus stars, para mergearlo con el df

```python
review = pd.read_csv("review.csv" ,usecols=["business_id", "stars"]) 
review.head(5)
```

```python
review["business_id"].value_counts()
```

Lo que quiero mergear es el mean de cada negocio, lo hago con groupby

```python
grouped = review.groupby("business_id").agg({"stars":"mean"})
```

```python
grouped.describe()
```

```python
grouped
```

```python
merged = dfCompleto.merge(grouped, left_on="business_id", right_on="business_id" , how= "inner")
merged
```

```python
plt.figure(dpi=150)
sns.violinplot(data = merged, x="name", y="stars" , palette=["#DFBD20","#248148","#653782","#265484","#C42D48"])
plt.title("Cual es la mejor cadena de comidas rapidas?")
plt.xlabel("Cadena")
plt.xticks(rotation=90)
plt.ylabel("Promedio de stats por local")
```
