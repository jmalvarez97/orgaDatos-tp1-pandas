```python
import pandas as pd
```

```python
df = pd.read_csv("ATP.csv")
```

```python
df
```

```python
df.columns
```

```python
short = df[["tourney_name", "surface", "tourney_level", "tourney_date", "winner_name" , "winner_ioc" , 'loser_name', "loser_ioc" , 'score']]
```

```python
short
```

```python
clay = short[short["surface"] == "Clay"]
clay
```

```python
clay["Year"] = pd.to_datetime(clay["tourney_date"], format='%Y%m%d').dt.year
```

```python
clay = clay[clay["Year"] < 2019]
clay
```

```python
argClay = clay[clay["winner_ioc"] == "ARG"]
argClay
```

```python
argClay["winner_name"].value_counts().head(20)
```

```python
argClay["winner_name"].value_counts().head(20).sum()
```

```python
def auxFunc(year):
    if 1698<=year<=1983:
        return "Era Vilas"
    elif 1984<=year<=1999:
        return "84-99"
    elif 2000<=year<=2010:
        return "Era La Legion"
    else:
        return "Era Davis"
```

```python
argClay["era"] = argClay["Year"].apply(lambda x: auxFunc(x))
```

```python
argClay
```

```python
argClay["era"].value_counts()
```

```python
eraVilas = argClay["era"] == "Era Vilas"
vilas = argClay[eraVilas]
vilas["winner_name"].value_counts()
```
